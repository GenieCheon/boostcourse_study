```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df[df["Year"] >= 2011]
```

```python
df[df["Year"] >= 2011].groupby(['Year','Country'])['Life_Expectancy'].mean().unstack('Country')
```

```python
df_2011 = df[df["Year"] >= 2011]
df_2011.pivot_table(values = "Life_Expectancy",
               index = "Year", 
               columns = "Country")
```
